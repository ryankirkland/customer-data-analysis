```python
state_df = grouped
state_df
```

```python
state_df = state_df.reset_index().rename(columns={'ship-state-abbr': 'State', 'shipment-id': 'Orders Count'})
```

```python
orders_by_state_df = state_df[['State', 'Orders Count']]
```
